# MovieLens 1m Data Set
- __Download and extract the moview from this site (http://www.grouplens.org/node/73)__
- __Download ml-1m__
- __It contain 1m dataset__
- __It spread across 3 tables rating, userinformation, and movie information__
- __*******************************__
- __Here users table is loaded into the pandas DataFrame object using pd.read_table__
- __read_table =>  read general delimited file into DataFrame__

```python
import pandas as pd
unames = ['user_id','gender','age','occupation','zipunames']
user = pd.read_table('ml-1m/users.dat',sep="::", names = unames)
user[:10]
```

- __Now load rating table into pandas DataFrame object using pd.read_table__

```python
import pandas as pd
rnames = ['user_id','movie_id','rating','timestamp']
rating = pd.read_table('ml-1m/ratings.dat', sep='::', header=None, names=rnames)
users[:10]
```

- __Now load movie name into pandas object__

```python
mnames = ['movie_id', 'title', 'genres']
movie = pd.read_table('ml-1m/movies.dat', sep='::', header=None, names=mnames)
movie[:10]
```

# If you want to analyze the data from more than one table which are associated with each other then we need to merged the data together into single table.

- __To merge the table we have 'merge' function of pandas__
- __'merge' function merge the tables__
- __At first we will merge rating with user, then merge the reuslt with the movie table__



```python
merge_data = pd.merge(pd.merge(user, rating), movie)
merge_data
```

- __Now let's find the mean rating of movie gourp by gender__
- __For this we can use the  'pivot_table' method of pandas__

```python
mean_rating = merge_data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_rating[:10]
```

- __Now let's find the movie that received at least 250 reviews__
- __'size' function in pands is use to get the total number of elements in the object__



```python

rating_by_title = merge_data.groupby('title').size()
# rating_by_title => have the movie with total rating, now let's find who have more than 250 review

active_rating = rating_by_title.index[rating_by_title>=250]
active_rating[:10]
```

#### Note
- __groupby() function is used to split the data into groups based on some criteria. pandas objects can be split on any of their axes. The abstract definition of grouping is to provide a mapping of labels to group names__
- __Indexing in pandas means simply selecting particular rows and columns of data from a DataFrame__

- __The index of titles receiving at least 250 ratings can then be used to select rows from mean_ratings above__

- __'loc' function is label-based, which means that we have to specify the name of the rows and columns that we need to filter out__

```python
mean_rating_at_least_250_reviews = mean_rating.loc[active_rating]
mean_rating_at_least_250_reviews
```

- __Top moviews among female viewrs__
- __The sort_values() function is used to sort by the values along either axis__


```python
top_female_ratings = mean_rating_at_least_250_reviews.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

# Measuring rating disagreement
- __Let's find the devisive male and female viewers__ 
- __One way is to add the column in mean_ratings containing the mean difference then sort by that__

```python
mean_rating_at_least_250_reviews['diff'] = mean_rating_at_least_250_reviews['M'] - mean_rating_at_least_250_reviews['F']
sorted_by_diff = mean_rating_at_least_250_reviews.sort_values(by='diff')
sorted_by_diff[:10]
```

- __Reversesing the order of rows, then slicing the top 10__
- __From this we get the movies prefered by men but women did not rate high__


```python
sorted_by_diff[::-1][:10]
```

#### Movie that is most disagreement between viewers

- __Disagreement can be measured by the standard deviation__
- __with groupby use (), to filter use []__
- __Filter


```python
rating_by_std = merge_data.groupby('title')['rating'].std()

# Now filter down by active ratings (which have more than 250 reviews)
rating_by_std_active_title = rating_by_std.loc[active_rating]
rating_by_std_active_title[:10]
```
