# Figures and Subplots

- __plots in matpltlib reside within a Figure object. we can create a new figure with plt.figure.__

- __You can’t make a plot with a blank figure. You have to create one or more subplots
using add_subplot__
- __This means that the figure should be 2 × 2 (so up to four plots in total), and we’re
selecting the first of four subplots (numbered from 1,2,3)__

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

- __'plt.subplots' create a new figure and returns a NumPy array containing the create subplot objects__

```python
fig,axes = plt.subplots(2,3)
axes
```

# Colors, Markers, and Line Styles
- __Example__

```python
plt.plot(randn(30).cumsum(), color='r', linestyle='dashed', marker='o')
```

```python
plt.plot(randn(30).cumsum(), color='r', linestyle='dashed', marker='o',label='Default')
plt.plot(randn(30).cumsum(), color='g', drawstyle='steps-post', label='Step plot')
plt.legend(loc='best')
```

# Ticks, Labels, and Legends
- __Setting the title, axis labels, ticks, and ticklabels__

- __example__

```python
fig = plt.figure()
ax  = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(), color='g')

#now let's set tile, axis
ticks  = ax.set_xticks([0, 250, 500, 750, 1000])
lables = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], fontsize='small', rotation='15')
ax.set_title('My first matplotlib')
ax.set_xlabel('Stages')
```

# Annotations and Drawing on a Subplot

- __Matplotlib have a object that represent many common shapes, refered as 'patches', like 'Rectangle', 'Circle' they are found in 'matplotlib.pyplot'__
- __But full set is located in 'matplotlib.patches'__

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='r')
circ = plt.Circle((0.7, 0.2), 0.15, color='b')
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g')

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

# Bar graph

- __'plot.bar()' make vertical bar plots__
- __'plot.barh()' makes horizontal bar plots__
- __In these case Series or DataFrame index will be used as the 'x' or 'y' ticks__
- __'alpha' is opacity of color__

- __Let's draw vertical bar graph__

```python
data = pd.Series(randn(5), index=list('abcdf'))
data.plot.bar(color='g', alpha=0.7)
```

- __Let's draw horizental bar graph__

```python
data = pd.Series(randn(5), index=list('abcdf'))
data.plot.barh(color='y', alpha=0.7)
```

- __Bar graph of DataFrame data__

```python
df = pd.DataFrame(np.random.rand(6, 4),
                  index = ['one', 'two', 'three', 'four', 'five', 'six'],
                  columns = pd.Index(['A', 'B', 'C', 'D'], name = 'Genus')
                 )
df.plot.bar()
```

- __Let's draw horizental__

```python
df.plot.barh()
```

- __by passing 'stacked = True', by resulting in value in each row being stacked together__

```python
df.plot.bar(stacked=True)
```

# Scatter or Point Plots

- __Points plots or scatter plots is useful for examining the relationship between two dimensional data series__
- __for example, here we load macrodata dataset from the statsmodel project, select few variables, then compute log differnce__

```python
macro = pd.read_csv('files/macrodata.csv')
macro[:5]
```

```python
#select specific columns

data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
data
```

```python
trans_data = np.log(data).diff().dropna()
trans_data
```

```python
sns.regplot('m1', 'unemp', data=trans_data)

plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```

- __Seaborn have 'pairplot' function which is use to look at all the scatter plots among group of variables__
- __this is known as a pairs plot or scatter plot matrix__

```python
sns.pairplot(trans_data, diag_kind='kde')
```
