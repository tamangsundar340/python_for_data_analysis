## Counting time zone in python
- __now let's count the time zone from records__
- __to count just write => len(time_zone) it will return us length of time_zone dict__
- __here we will only phase data from 0 index to 10 index__

```python
time_zone = [rec['tz'] for rec in records if 'tz' in rec]
time_zone[0:10]
```

- __some of the datas are empty and unknown here__
- __let's create funtion to phase time_zone from the list__


```python
from collections import defaultdict

def count_zone(list):
    count = defaultdict(int)
    for i in list:
        count[i] += 1
    return count

count = count_zone(time_zone)
count
```

- __as we see in the above 'America/New_York' have 1251, and 521 in empty__
- __Now if you want a top 10 time zone with count then we have Counter of python library__
- __example:-__

```python
from collections import Counter
count = Counter(time_zone)
count.most_common(10)
```

## Decompose top time zone base on windows and non windows user, excluding the null values
- __at first we only select a value from 'a' column which do not have null values__
- __Then we checked whether 'a' columns row is Windows or not__
- __Then print first five values__

```python
import numpy as np
not_null_frame = frame[frame.a.notnull()]
os = np.where(not_null_frame['a'].str.contains('Windows'), 'Windows', 'Not windows')
os[:5]
```

- __At first we group by not_null_frame base on 'tz' and 'os'__
- __Then we computed by_tz_os by 'size' function, reshaped into table using 'unstack' function and fill 0 into missing values using 'fillna' function__

```python
by_tz_os = not_null_frame.groupby(['tz', os])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

### Let's select the top overal time zone
- __ I construct an indirect index array from the row counts in agg_counts__

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

- ___I then use 'take' to select the rows in that order, then slice off the last 10 rows__

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

- __Now let's plot this into the graph__

```python
count_subset.plot(kind='barh', stacked=True)
```
