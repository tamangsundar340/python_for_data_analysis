# Data Loading, Storage, and Files Formats

- __The cat command allows us to create single or multiple files, view contain of file, concatenate files and redirect output in terminal or files__

```python
# cat doe not provide us the heading

data = !cat files/business.csv
data[:5]
```

```python
# read files using read_csv function
# read_csv funciton provide us heading

import pandas as pd

df = pd.read_csv('files/business.csv')
df[:5]
```

```python
# if we put header as None then read_csv function also do not provide us header

df_2 = pd.read_csv('files/business.csv', header=None)
df_2[:3]
```

```python
# use names with read_csv

df_3 = pd.read_csv('files/business.csv', names=['a','b','c','d','e','f','g','h','i','j','k','l','m','n'])
df_3[:3]
```

```python
# we could also used read_table function and use delimete with it

df_5 = pd.read_table('files/business.csv', sep=',')
df_5[:3]
```

```python
result = pd.read_table('files/business.csv')
result[:3]
```

### Reading Text Files in Pieces

- __if you really want a small number of rows from large set then use like this__

```python
df_5 = pd.read_csv('files/business.csv',nrows=3)
df_5
```

- #### To read a file in pieces, specify a chunksize as a number of row

```python
chunker = pd.read_csv('files/business.csv', chunksize=1000)
chunker
```

- __Now we can iterate over the chunker size and store them into somewhere__

```python
to_series = pd.Series([])
to_series
```

```python
for i in chunker:
    to_series = i
to_series
```

#### Writing Data to Text Format

```python
data = pd.read_csv('files/business.csv')
data[:5]
```

 - __Using DataFrames to_csv method__

```python
data.to_csv('files/out.csv')
```

- __we have written into 'out.csv' files now let's read it__

```python
read_files = pd.read_csv('files/out.csv')
read_files[:3]
```

```python
# we can also use delimeter while writting into file
import sys

result = data.to_csv(sys.stdout, sep='|')
result
```

```python
# If there is any null value then fill it NULL

result_2 = data.to_csv(sys.stdout, na_rep='Null')
result_2
```

```python
# we can disabled both index and header 

result_3 = data.to_csv(sys.stdout, index=False, header=False)
result_3
```

```python
# write subset

result_4 = data.to_csv(sys.stdout, index=False, columns=['Series_reference', 'Period','Data_value'])
result_4
```

- __Another method let's read a data in the list of lines__

```python
with open('files/business.csv') as f:
    lines = list(pd.read_csv(f))
    print(lines)
```

- __Let's split into header and data line__

```python
header, value = lines[0], lines[1:]
header
value
```

- __Write into delimted files__

```python
with open('mydata.csv', 'w') as f:
    writter = csv.writer(f, dialect='-')
    writter.writerow('one', 'two', 'three')
    writter.writerow('1', '2', '3')
    writter.writerow('11', '22', '33')
```

#### JSON Data

- __To convert a JSON string to Python form, use json.loads :__

```python
obj = """
{"name": "Wes",
"places_lived": ["United States", "Spain", "Germany"],
"pet": null,
"siblings": [{"name": "Scott", "age": 30, "pets": ["Zeus", "Zuko"]},
{"name": "Katie", "age": 38,
"pets": ["Sixes", "Stache", "Cisco"]}]
}
"""
```

```python
import json

to_python_data = json.loads(obj)
to_python_data
```

```python
# phase json data usin the dataframe

result = pd.DataFrame(to_python_data['siblings'], columns=['age', 'name'])
result
```

#### 'to_json' method convert pandas data into JSON

```python
to_json_data = to_python_data.to_json()
to_json_data
```
