# Files and the Operating System

- __Most of this book uses high-level tools like pandas.read_csv to read data files from disk into Python data structure__
- __However, itâ€™s important to understand the basics of how to work with files in Python__

```python
# Example to work with files

path = 'names/yob1880.txt'
f = open(path)

for line in f:
    print(line)
    
# or we do this 

for line in open('names/yob1880.txt'):
    print(line)
```

```python
# Or we can do this

path = 'names/yob1880.txt'
to_line = [line.upper() for line in open(path)]
to_line
```

- __When we use open to create an object, it is important to cloe the file when finished__

```python
f.close();
```

- __We dont' need to write 'f.close()' when we do this__

```python
with open(path, 'r') as f:
    lines = [x.rstrip() for x in f]
    print(lines)
```

### Notes
- __open(path, 'r') =>read only mode__
- __open(path, 'w') =>write only mode, erase previous file and write new__
- __open(path, 'x') =>write only mode, fail if same file already exist__
- __open(path, 'a') =>append to the file, create if not exist__
